```python
def texto_a_lista(texto):
    lista_resultado = texto.split('\n')
    lista_resultado = [item.strip() for item in lista_resultado if item]
    return lista_resultado

```

```python
def filtrar_columnas(dataframe, lista_nombres):
    """
    Filtra un DataFrame para incluir solo las columnas cuyos nombres están en la lista_nombres.

    Parameters:
    - dataframe: DataFrame de pandas.
    - lista_nombres: Lista de nombres de columnas a incluir.

    Returns:
    - Nuevo DataFrame con las columnas seleccionadas.
    """
    nuevo_dataframe = dataframe[lista_nombres]
    return nuevo_dataframe
```
